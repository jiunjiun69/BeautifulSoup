```python
import requests
from bs4 import BeautifulSoup 
```

```python
url = "https://www.ptt.cc/bbs/Beauty/index.html"
response = requests.get(url)
print(response.text)
```

```python
session = requests.Session()
session.post('https://www.ptt.cc/ask/over18', data = {'from' : '/bbs/Beauty/index.html', 'yes' : 'yes'})
response = session.get(url)
soup     = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
selects = soup.select("div.title a") #HTML中的div class="title"中的a標籤
```

```python
for select in selects:
    print(select["href"], select.text) 
```
